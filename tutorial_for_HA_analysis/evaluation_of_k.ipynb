```r
ks <- c(50,100,200,500,1000)
m <- 100

As <- list()
for(i in 1:length(ks)){
  k <- ks[i]
  As[[i]] <- as.matrix(read.table(paste0("data/A_k",k,".txt")))
}

X <- as.matrix(read.table("data/X.txt", header=TRUE))
for(j in 1:dim(X)[2]){
    X[,j] <- X[,j] / sqrt(sum(X[,j]**2))
}
```

```r
for(i in 1:length(ks)){
    k <- ks[i]

    #construct graph Laplacian
    L <- sym_normalized_graph_laplacian(As[[i]])
    
    #perform GFT
    tmp <- graph_fourier_transform(L,X,m)
    U <- tmp$U
    eigenvalues <- tmp$eigenvalues
    hF <- tmp$hF
    tF <- tmp$tF

    plot_aGLR(X, U, 100)
}
```
